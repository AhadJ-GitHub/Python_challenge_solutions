**Exercise 2:** MÃ¼xtÉ™lif siniflÉ™rin cinsÉ™ vÉ™ yaÅŸ aralÄ±qlarÄ±na gÃ¶rÉ™ saÄŸ qalma sayÄ±nÄ± gÃ¶stÉ™rÉ™n pivot cÉ™dvÉ™li qurun

Qeyd: YaÅŸ kateqoriyalarÄ± 0 - 10; 10 - 30; 30 - 60; 60 - 80 vÉ™ yaÅŸÄ± mÃ¼É™yyÉ™n edilmÉ™yÉ™nlÉ™r ÅŸÉ™kilindÉ™ olmalÄ±dÄ±r.

```python
# your code
ara=pd.cut(tnc['Age'],[0,10,30,60,80])
ara=ara.cat.add_categories('Age is NaN').fillna('Age is NaN')
surv=tnc.pivot_table(values='Survived',index=['Sex',ara],columns='Pclass',aggfunc='count',dropna=False)
surv
```

**Exercise 3:** Titanikin sÉ™rniÅŸinlÉ™rini É™vvÉ™lki Ã§alÄ±ÅŸmadakÄ± kimi yaÅŸ aralÄ±qlarÄ±na gÃ¶rÉ™ qruplaÅŸdÄ±rÄ±n vÉ™ hÉ™r qrupda A kabinindÉ™ olan sÉ™rniÅŸinlÉ™rin sayÄ±nÄ±, maximum vÉ™ minimum yaÅŸ fÉ™rqlÉ™rini É™ks etdirin

```python
# your code
dts0=tnc[tnc['Cabin'].str.upper().str.startswith('A')==True]
ara2=pd.cut(dts0['Age'],[0,10,30,60,80])
ara2=ara2.cat.add_categories('Age is NaN').fillna('Age is NaN')
myF=lambda x: x.max()-x.min()
dts1=dts0.pivot_table(index=ara2,aggfunc={'Age': myF,'Cabin':'count'})
dts1.rename(columns={'Age':'Max-Min in Age','Cabin':'Total quantity'},inplace=True)
dts1.columns.names=['All in Cabin A']
dts1
```

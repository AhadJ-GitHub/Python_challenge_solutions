**Exercise 1:** Müxtəlif siniflərin cinsə görə sağ qalma nisbətini tapmaq üçün pivot cədvəli qurun

```python
# your code
import numpy as np
result = tnc.pivot_table(values='Survived', index='Sex', columns='Pclass')
result
```

**Exercise 2:** Müxtəlif siniflərin cinsə və yaş aralıqlarına görə sağ qalma sayını göstərən pivot cədvəli qurun

Qeyd: Yaş kateqoriyaları 0 - 10; 10 - 30; 30 - 60; 60 - 80 və yaşı müəyyən edilməyənlər şəkilində olmalıdır.

```python
# your code
ara=pd.cut(tnc['Age'],[0,10,30,60,80])
ara=ara.cat.add_categories('Age is NaN').fillna('Age is NaN')
surv=tnc.pivot_table(values='Survived',index=['Sex',ara],columns='Pclass',aggfunc='count',dropna=False)
surv
```

**Exercise 3:** Titanikin sərnişinlərini əvvəlki çalışmadakı kimi yaş aralıqlarına görə qruplaşdırın və hər qrupda A kabinində olan sərnişinlərin sayını, maximum və minimum yaş fərqlərini əks etdirin

```python
# your code
dts0=tnc[tnc['Cabin'].str.upper().str.startswith('A')==True]
ara2=pd.cut(dts0['Age'],[0,10,30,60,80])
ara2=ara2.cat.add_categories('Age is NaN').fillna('Age is NaN')
myF=lambda x: x.max()-x.min()
dts1=dts0.pivot_table(index=ara2,aggfunc={'Age': myF,'Cabin':'count'})
dts1.rename(columns={'Age':'Max-Min in Age','Cabin':'Total quantity'},inplace=True)
dts1.columns.names=['All in Cabin A']
dts1
```

**Exercise 4:** Hər sinifdə sağ qalan qadın sərnişinlərin sayını tapın və onu yeni bir sütun olaraq Titanic cədvəlinə əlavə edin 

```python
# your code
tnc2=tnc.copy()
srs=tnc2[(tnc2['Sex']=='female') & (tnc2['Survived']==1)].groupby(by='Pclass').count()['Survived']
tnc2['# of Surv. fem. in class']=tnc['Pclass'].map(srs)
tnc2
```

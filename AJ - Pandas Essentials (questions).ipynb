```python
# Question 10
# Do females tip more than males? Which measures do you use for this problem and why? (tips data)
tips__=pd.read_csv('tips.csv')
tips__.drop('Unnamed: 0', axis=1, inplace=True)
tips__
```

```python
tips__.groupby(by='sex').sum()['tip']
# In total, for this dataset, males have tipped more the females.
```

```python
tips__.groupby(by='sex').count()['tip']
# However, the number of males is higher than females
```

```python
tips__.groupby(by='sex').mean()['tip']
# This shows the average amount of tips given by females and males.
# Here, we can tell that males tip more than females.

# However, ...
```

```python
sns.scatterplot(y=tips__['tip'], x=tips__['total_bill'])
```

```python
sns.scatterplot(y=tips__['tip'], x=tips__['total_bill'], hue=tips__['sex'])
```

```python
# Here, we can observe that the more the total bill is, the more the person tips.
# This means that the simple average would be biased because the total bills differ.
# This leads us to an idea to take the average weighted by the total bill.
```

```python
# So, here we go again :D

values_f = tips__[tips__['sex']=='Female']['tip']
weights_f = tips__[tips__['sex']=='Female']['total_bill']
ave_f=round(np.average(values_f, weights=weights_f),2)

values_m = tips__[tips__['sex']=='Male']['tip']
weights_m = tips__[tips__['sex']=='Male']['total_bill']
ave_m=round(np.average(values_m, weights=weights_m),2)

new_stats=pd.DataFrame({'Female':ave_f,'Male':ave_m},index=[0])
new_stats

# Based on these results, males still tend to tip more than female.
# However, the categorical columns such as smoker, day, time may also be related to the amount of tip given.
# There is a need for an extended analysis. :D
```

```python
# Question 1
# Read both car datasets and drop the unnamed columns on cars1. Then unite these
# datasets rowise. Then create a random number of Series from 15,000 to 73,000 and add
# it to the cars dataset under name of 'owners'. It will be a new column (cars1, cars2 data)
import numpy as np
import pandas as pd
```

```python
cars1=pd.read_csv('cars1.csv')
cars2=pd.read_csv('cars2.csv')

cars1.drop(cars1.columns[cars1.columns.str.startswith('Unnamed:')],axis=1,inplace=True)
newcars=pd.concat([cars1,cars2],ignore_index=True)
randnums=pd.Series(np.random.randint(15000,73000.1,size=(398,)))
final=pd.concat([newcars,randnums],axis=1)
final.rename(columns={0:'owner'},inplace=True)
final
```

```python
# Question 2
# For each combination of occupation and gender, calculate the mean age 
# It would be great if you'd show the results in a DataFrame format (Users data)
users=pd.read_csv('Users.csv')
users.groupby(by=['gender','occupation']).mean()[['age']]
```

```python
# Question 3 
# For each occupation present the percentage of women and men (Users data)
groupped=users.groupby(['occupation','gender'])['age'].count()
users2=groupped.rename("percentages").groupby(level = 0).transform(lambda x: x/x.sum())
users2=users2*100
users2=pd.DataFrame(users2)
users2
```

#### Exercise 3

Bu gün, dünən və sabahkı tarixləri çap edən proqram yazın. Tarixləri (məsələn) `08 December, 2022` formatında çıxarın

```python
# your code

today_ = pd.to_datetime(datetime.datetime.now())
yesterday_ = pd.to_datetime(today_ - datetime.timedelta(days=1))
tomorrow_ = pd.to_datetime(today_ + datetime.timedelta(days=1))

print(f"Today's date is {today_.strftime(format='%d %B, %Y')}")
print(f"Yesterday's date is {yesterday_.strftime(format='%d %B, %Y')}")
print(f"Tomorrow's date is {tomorrow_.strftime(format='%d %B, %Y')}")

# Alternatively,

# print(f"Today's date is {today_.day} {today_.month_name()}, {today_.year}")
# print(f"Yesterday's date is {yesterday_.day} {yesterday_.month_name()}, {yesterday_.year}")
# print(f"Tomorrow's date is {tomorrow_.day} {tomorrow_.month_name()}, {tomorrow_.year}")
```

#### Exercise 5

İstifadəçidən ay(numeric) və il(numeric) daxil etməsini istəyin verilən ilə görə aylıq təqvim çap edin (tarix və həftənin günləri ilə) 

```python
# your code
while True:
    
    print('\nPress q to exit\n')
    
    year_=input('Enter the year in the format YYYY:')
    
    if year_=='q':
        break
    else:
        month_=input('Enter the month in the format MM:')
    
    teqvim=pd.Period(f'{year_}-{month_}', freq='M')
    calendar=pd.date_range(teqvim.start_time,teqvim.end_time)
    
    calendar2=pd.DataFrame()
    calendar2['days of week']=calendar.day_name()
    calendar2['dates']=calendar
    
    display(calendar2)
```
